# **CM1**

```python
df1 = pd.read_json("drive/MyDrive/TonicDataset/CM/Annotations/CM1.json")
length = 271

metadata = pd.concat([df1.transpose()])
metadata = metadata.transpose()

metadata = metadata.drop("artist")
metadata = metadata.drop("gender")
metadata = metadata.drop("type")
metadata = metadata.drop("tradition")
```

```python
worksheet = gc.open('AAR AF Spec').sheet1
spec_obj = af.MelSpectrogram(num=64, samplate=44100, radix2_exp=19)

ct=1

for column in metadata:
  path = column
  # Read audio data and sample rate
  audio_arr, sr = af.read("drive/MyDrive/" + path)
  # X = audio time series, cc_num = # of MFCCs, mel_num = default # of frequency bins, fft_length = 2**radix2_exp
  spec_arr = spec_obj.spectrogram(audio_arr)
  print(spec_arr.shape)
  break
  # slide_length (hop length) = fft_length/4 -- see documentation https://audioflux.top/_autosummary/audioflux.mfcc.html#audioflux.mfcc
  spec_arr = spec_arr.transpose()
  spec_arr = spec_arr.flatten()
  spec_arr = spec_arr.tolist()
  data = pd.DataFrame(spec_arr)
  data = data.transpose()
  #path = str(ct)+'.csv'
  #data.to_csv(path)
  #files.download(path)
  count_col = data.shape[1]
  cell_list = worksheet.range(ct,2,ct,1+count_col)
  for i in range(0, count_col):
    cell_list[i].value = data.loc[0,i]
  ct += 1
  worksheet.update_cells(cell_list)
```

```python
df1 = pd.read_json("drive/MyDrive/TonicDataset/CM/Annotations/CM3.json")
length = 428

metadata = pd.concat([df1.transpose()])
metadata = metadata.transpose()

metadata = metadata.drop("artist")
metadata = metadata.drop("gender")
metadata = metadata.drop("type")
metadata = metadata.drop("tradition")
```

```python
import math
worksheet = gc.open('AAR AF Spec').sheet1
ct=1207

for column in metadata:
  path = column
  # Read audio data and sample rate
  audio_arr, sr = af.read("drive/MyDrive/" + path)
  # X = audio time series, cc_num = # of MFCCs, mel_num = default # of frequency bins, fft_length = 2**radix2_exp
  spec_obj = af.MelSpectrogram(num=64, samplate=44100, radix2_exp=19, slide_length=745*math.ceil(lb.get_duration(y=audio_arr,sr=sr)))
  spec_arr = spec_obj.spectrogram(audio_arr)
  if spec_arr.shape[1]>57: # remove last few "frames" if greater than 57 "frames"
    spec_arr = spec_arr[ : , 0 : 57]
  # slide_length (hop length) = fft_length/4 -- see documentation https://audioflux.top/_autosummary/audioflux.mfcc.html#audioflux.mfcc
  spec_arr = spec_arr.transpose()
  spec_arr = spec_arr.flatten()
  spec_arr = spec_arr.tolist()
  data = pd.DataFrame(spec_arr)
  data = data.transpose()
  #path = str(ct)+'.csv'
  #data.to_csv(path)
  #files.download(path)
  count_col = data.shape[1]
  cell_list = worksheet.range(ct,2,ct,1+count_col)
  for i in range(0, count_col):
    cell_list[i].value = data.loc[0,i]
  ct += 1
  worksheet.update_cells(cell_list)
```
